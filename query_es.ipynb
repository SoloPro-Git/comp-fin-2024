```python
from elasticsearch import Elasticsearch
import pandas as pd

# 连接到ES
# 请替换为你的实际ES连接信息
es = Elasticsearch(
    ['http://localhost:9200'],
    basic_auth=('elastic', 'elastic')  # 添加认证信息
)

def get_index_mappings(index_pattern="表字段信息"):
    # 构建查询
    # 构建搜索公司名字的查询
    query = {
        "query": {
            "match": {
                "column_description": "中文名字"
            }
        }
    }
    
    # 执行查询
    try:
        results = es.search(
            index=index_pattern,
            body=query
        )
        
        # 提取结果
        hits = results['hits']['hits']
        
        # 转换为DataFrame
        records = []
        for hit in hits:
            records.append(hit['_source'])
            
        return pd.DataFrame(records)
        
    except Exception as e:
        print(f"查询出错: {str(e)}")
        return None
# 执行查询并显示结果
df = get_index_mappings()
if df is not None:
    display(df)
else:
    print("未找到结果")

```
